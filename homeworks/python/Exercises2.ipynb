Plot a rainbow

```python
from matplotlib import pyplot as plt
import numpy  as np
import math as math
```

```python
x = np.linspace(0, math.pi, 100)

N = 30
for i in range(0, N):
    y = np.sin(x)- 0.1*i
    plt.plot(x, y, color=(255/N*i/255, 255/N*i/2555, 255/N*i/255))


```

### **Exercise2:**

Plot the recaman sequence in the following way:

Add a semi-circular arc from $a_{n}$ to $a_{n-1}$.

Alternate the arcs to be above and below the x-axis according to whether
    n is even or odd.

```python
def recamanSequence(size):
    sequence = []
    visited = set()

    visited.add(0)
    sequence.append(0)

    n = 1
    while(n <= size):
        if (sequence[-1] - n) > 0 and (sequence[-1] - n) not in visited:
            visited.add(sequence[-1] - n)
            sequence.append(sequence[-1] - n)
        else:
            visited.add(sequence[-1] + n)
            sequence.append(sequence[-1] + n)
        n += 1
    return sequence

print(recamanSequence(10))
```

```python
from matplotlib.patches import Arc
```

```python
N = 50
seq = recamanSequence(N)
print(seq)
maxValInSeq = max(seq)

fig, ax = plt.subplots()
ax.set_xlim(0, maxValInSeq * 1.05)
ax.set_ylim(-1, 1)

# Axes
ax.axhline(0.0)

for i in range(0, N):
    length = math.fabs(seq[i]-seq[i+1])
    if i % 2 == 0:
        ax.add_patch(Arc(( (seq[i]+seq[i+1])/2, 0), length, 1,
                         theta1=0, theta2=180, edgecolor='k', lw=1.5))
    else:
        ax.add_patch(Arc(( (seq[i]+seq[i+1])/2, 0), length, 1,
                         theta1=180, theta2=0, edgecolor='k', lw=1.5))




plt.show()
```

### **Exercise3:**

Write a matrix multiplication fucntion using lists ("manually", like you did in c++) and another one using numpy arrays (call matmul function). Time them both.

```python
import numpy as np
import random as rd
```

```python
def generate_random_square_matrix(n:int) -> list:
    output = list()
    for i in range(0, n):
        output.append([])
        for j in range(0, n):
            output[i].append(rd.randint(-1000, 1000))
    return output


def matrixMultiplicationWithList(mat1, mat2, N):
    # create output matrix
    output = list()
    for i in range(0, N):
        output.append([0]*N)

    for i in range(0, N):
        for j in range(0, N):
            temp = 0
            for k in range(0, N):
                temp += mat1[i][k] * mat2[k][j]
            output[i][j] = temp
    return output



N = 80
mat1 = generate_random_square_matrix(N)
mat2 = generate_random_square_matrix(N)

%%timeit
result1 = matrixMultiplicationWithList(mat1, mat2, N)



#input
nparray1 = np.array(mat1)
nparray2 = np.array(mat2)


result2 =  np.matmul(nparray1, nparray2)

```

```python
%%timeit
result1 = matrixMultiplicationWithList(mat1, mat2, N)
```

```python
%%timeit
result2 =  np.matmul(nparray1, nparray2)
```
